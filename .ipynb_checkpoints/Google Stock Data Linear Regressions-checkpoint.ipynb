```python
# data found at:  https://finance.yahoo.com/quote/GOOG/history?p=GOOG\

'''
This data is purely stock amounts for Alphabet. 

In this notebook I experimented with some linear regressions after building some features.

'''
```

```python
import pandas as pd
import os
import warnings
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn import linear_model

filename = 'C://Users/Tim/Downloads/GOOG.csv'

dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d')
dataDF = pd.read_csv(filename, index_col='Date', parse_dates=['Date'], date_parser=dateparse)
```

```python
y_name = 'Close'
dataDF.plot(y = 'Close')
```

```python
#result = seasonal_decompose(df_close, model='multiplicative', freq = 30)
df_close = dataDF['Close']

result = seasonal_decompose(df_close, model='multiplicative', freq = 15)
fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(16, 9)
```

```python
trend = result.trend.dropna()
trend
```

```python
closeDF = pd.DataFrame(df_close) #(trend) # (close) 
closeDF.columns = ['Close']
```

```python
def feature_gen(df):
    df['roll_avg_10days'] = closeDF['Close'].rolling(50).mean()
    df['roll_avg_5days'] = closeDF['Close'].rolling(10).mean()
    
    df = df.dropna()
    return df
```

```python
featDF = feature_gen(closeDF)
```

```python
from datetime import timedelta
featDF['date_shift'] = featDF.index + timedelta(days = 7)

label_col = featDF[['Close']]
label_col.columns = ['label'] 
label_col['date_shift'] = label_col.index

featDF['Date'] = featDF.index
featDF = pd.merge(featDF, label_col, how = 'inner', on = 'date_shift')
#featDF = featDF.set_index('Date')
featDF.index = featDF['Date']

cols_wanted = [c for c in featDF.columns if (c not in ['Date', 'date_shift'])]

featDF = featDF[cols_wanted]

featDF

```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10, 6

plt.legend(loc='best')
plt.title('Moving Average')
plt.plot(featDF['Close'], color ="black", label = "Close")
plt.plot(featDF['roll_avg_10days'], color="red", label = "roll_avg_10days")
plt.legend()
plt.show()
```

```python
part = 0.7
train_data, test_data = featDF[3:int(len(featDF)*part)], featDF[int(len(featDF)*part):]
```

```python
import statsmodels.api as sm

model = sm.OLS(train_data[['label']], train_data.drop(columns = ['label']))
res = model.fit()

```

```python
print ( res.summary())
```

```python
predict = res.predict( test_data.drop(columns = ['label']))

predict
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10, 6

original_label = df_close[int(len(featDF)*part):][test_data.index]

plt.legend(loc='best')
plt.title('Moving Average')
plt.plot(test_data['label'], color ="black", label = "label")
plt.plot(predict, color="red", label = "predict")
plt.plot(original_label, color='green', label = 'original')
plt.legend()
plt.show()
```

```python
residuals = abs(predict - test_data['label'])

plt.plot(residuals, color = 'black')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

labels = test_data['label']

# report performance
mse = mean_squared_error(labels, predict)
print('MSE: '+str(mse))
mae = mean_absolute_error(labels, predict)
print('MAE: '+str(mae))
rmse = math.sqrt(mean_squared_error(labels, predict))
print('RMSE: '+str(rmse))
mape = np.mean(np.abs(predict - labels)/np.abs(labels))
print('MAPE: '+str(mape))
```
